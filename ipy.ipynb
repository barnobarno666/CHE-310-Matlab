```python
%%octave  
T_in=  60 ;
T_end  = 70; 
d= 0.0254 ; 
v = 0.02 ; 
u =  0.000471  ;
T2 = 70 ;  
T2_temp = 60 ; 
Tw= 80  ; 
OUTLET = [] ; 
INLET = [] ;
x=[] ; 
for  T =  T_in : 2 : T_end 
    for i = 1 : 1 : 10000 
        Tm=(T+T2_temp)/2 ;
        density =  -0.4067 * Tm +1000.8 ;
        Re = d*v*density /(u);
        h= 14.6167 * Re^(1/3);
        m= density *1.013*10^-5 ;
        T2 = (h * 0.2394 * (Tw-0.5*T) + 4180 * m *T ) /(4180*m + 0.1197 * h) ;
        if abs(T2_temp-T2) < 0.001
            disp(i)
            %disp('ass')
            break
        end
        T2_temp = T2 ;
    end
    
    INLET = [INLET T] ;
    OUTLET = [OUTLET T2];
end

disp(OUTLET) 
disp('ASS')   
    
    
 
 




```

```python
%%octave
Rex = [ ];
Re = [ ] ;
for  i = 1 : 1 :10
    Rex = [] ;
    for j = 0.5 : 0.5 : 5 
        R = i* j * 1000*10^-3 / (1.002 * 10^-3) ; 
        Rex =[Rex  R] ; 
    end
    Re = [Re ; Rex] ;
    if i==1 
    disp(Re)
    end 
end
```

```python
%%octave 
const = 1000e-3 / (1.002e-3);  % Precompute constant
i_vec = 1:10;  % Row indices (i)
j_vec = 0.5:0.5:5;  % Column indices (j), 10 values
[I, J] = meshgrid(i_vec, j_vec);  % I: 10x10 with rows as i, J: 10x10 with columns as j
Re = I .* J * const;  % Element-wise multiplication
disp(Re);  % Display the full matrix
```
