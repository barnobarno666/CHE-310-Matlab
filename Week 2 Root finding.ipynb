```python
%%octave
number  =1000
S =  @(x) (x.^2 -number) ;
b = [1:1:10000] ;
a=[1:-1:-10000]; 
a= min(a(S(a) < 0));
b=max(b(S(b)>0)) ;
MAX_ITER = 100_000
iter = 0
while true
    c=(a+b)/2 ;
    if S(a) * S(c) <0 
        b=c ;
    elseif S(b) * S(c) < 0 
        a=c ;
    else 
        break
    end 
    iter+=1 ;
    if iter>MAX_ITER 
        break
    end 
end
disp(c)
```

```python
%%octave 
function c = ROOT(number, ITER)
    if nargin < 2 
        ITER =10_000 ;
    end
    S = @(x) (x.^3 .-number); 
    a= [0:-1:-number] ;
    b= [1: 1: number];
    a=min(a(S(a) < 0)) ;
    b= max(b(S(b) > 0));
    for i = 0 : 1: ITER 
        c=(a+b)/2 ; 
        if S(a)*S(c) < 0 
            b = c ;
        elseif S(c) *S(b) <  0 
            a = c ;
        else
            disp(["Ended at iteration: " , num2str(i) ]  )
            break
        end         
    end  
end
root=ROOT(1000 , 100)
```
